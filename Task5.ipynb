```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv', encoding = "ISO-8859-1", low_memory=False)
data.head()
```

```python
pd.options.display.max_columns = None
data.sample(6)
```

```python
data.columns.values
```

```python
data_for_analysis = data.loc[:,['iyear', 'imonth', 'iday','country_txt', 'region_txt','city', 'latitude', 'longitude', 
'attacktype1_txt','targtype1_txt','target1','targsubtype1_txt','corp1','gname','motive','weaptype1_txt','nkill','nwound','success','summary']]
data_for_analysis.sample(5)
```

```python
def fill_zeros(columns):
    for i in columns:
        data_for_analysis[i] = np.where(data_for_analysis[i] == 0,1, data_for_analysis[i])
fill_zeros(['imonth','iday'])
```

```python
data_for_analysis['date'] = [(str(data_for_analysis['iyear'][i]) + '-' + str(data_for_analysis['imonth'][i]) +'-' + str(data_for_analysis['iday'][i])) for i in np.arange(data_for_analysis['iday'].shape[0]) ]
data_for_analysis['date']
```

```python
data_for_analysis['date'] = pd.to_datetime(data_for_analysis['date'])
```

Question no. 1: what was the most devastating terrorist attack each year?

```python
year_nkill =data_for_analysis.loc[:,['iyear','nkill']].groupby('iyear').max()
```

```python
most_kills = pd.DataFrame(columns=data_for_analysis.columns.values)
```

```python

for index, row in year_nkill.iterrows():
    most_kills = pd.concat([most_kills,data_for_analysis[(data_for_analysis['iyear'] == index) & (data_for_analysis['nkill'] == row[0])]])
```

```python
most_kills.sample(5)
```

Question no.2 : What year had the most terror attacks attempts/ victims?

```python
most_kills[most_kills['nkill'] == most_kills['nkill'].max()]
```

Question no.3 : 3. what is the country/ target with the most terrorist attacks/ victims?

```python
country_attacks = data_for_analysis.pivot_table(index='country_txt', columns='iyear', aggfunc='count')
```

```python
country_attacks['attacktype1_txt'][1970][country_attacks['attacktype1_txt'][1970] == country_attacks['attacktype1_txt'][1970].max()]
```

```python

country_dict = {}
for i in range(1970,2018):
    try:
        dict_series = country_attacks['attacktype1_txt'][i][country_attacks['attacktype1_txt'][i] == country_attacks['attacktype1_txt'][i].max()]
        country_dict[i] = [dict_series.index[0],dict_series[0]]
    except KeyError: continue
pd.DataFrame(country_dict)
```

4. what is the most used attacking method each year?

```python
type_attacks = data_for_analysis.pivot_table(index='attacktype1_txt', columns='iyear', aggfunc='count')
```

```python

type_dict = {}
for i in range(1970,2018):
    try:
        dict_series = type_attacks['city'][i][type_attacks['city'][i] == type_attacks['city'][i].max()]
        type_dict[i] = [dict_series.index[0],dict_series[0]]
    except KeyError: continue
pd.DataFrame(type_dict)
```

Question no.5: which terrorist group had the most terrorist attacks attempts?

```python
terror_group = data_for_analysis.pivot_table(index='gname', columns='iyear', aggfunc='count')
```

```python
terror_group
```

```python
terror_group['attacktype1_txt'][1970][terror_group['attacktype1_txt'][1970] == (terror_group['attacktype1_txt'][1970]).max()]
```

```python

terror_dict = {}
for i in range(1970,2018):
    try:
        dict_series = terror_group['attacktype1_txt'][i][terror_group['attacktype1_txt'][i] == (terror_group['attacktype1_txt'][i]).max()]
        terror_dict[i] = [dict_series.index[0],dict_series[0]]
    except KeyError: continue
pd.DataFrame(terror_dict)
```

```python
list_of_top_groups = []
filter = data_for_analysis['gname'].value_counts()
for i in filter.index.values:
    if (filter[i] > 1) == True:
        list_of_top_groups.append(i)
```

```python
data_cleaned = data_for_analysis[data_for_analysis['gname'].isin(list_of_top_groups)]
```
